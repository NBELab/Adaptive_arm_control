```python
from IK import *

arm   = widow200()
q, trajectory, error_tract, xyz_t = goto_target(arm, np.array([[-0.25, 0.1, 0.1]]).T, optimizer = Optimizer.DLS)
```

```python
np.rad2deg(q)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

# Plot EE convergence to target -------------------------------------------------------
plt.figure()
plt.ylabel("Distance (m)")
plt.xlabel("Time (ms)")
plt.title("Distance to target")
plt.plot(error_tract)
plt.show()

# Plot EE trajectory ------------------------------------------------------------------
ax = plt.figure().add_subplot(111, projection='3d')
ee_x = [ee[0][0] for ee in trajectory]
ee_y = [ee[1][0] for ee in trajectory]
ee_z = [ee[2][0] for ee in trajectory]

ax.set_title("End-Effector Trajectory")
ax.plot(ee_x, ee_y, ee_z)
ax.scatter(xyz_t[0], xyz_t[1], xyz_t[2], label="target", c="r")
ax.legend()   
```
